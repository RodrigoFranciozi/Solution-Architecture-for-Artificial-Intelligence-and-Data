```
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```
X, y = make_blobs(n_samples=150, centers=3, cluster_std = 0.6, n_features=2,random_state=0)
plt.scatter(X[:, 0], X[:, 1], c='white', marker='o', edgecolor='black')
plt.show()
```

```
#Kmeans
from sklearn.cluster import KMeans

Km = KMeans( n_clusters = 3, init = 'random', max_iter = 100, random_state = 0, n_init = 10)
```

```
Km.fit(X)
```

```
Km.score(X)
```

```
Km.inertia_ #Quanto maior (centros e dados muito dispersos) pior, quanto menor melhor. 
```

```
y_km = Km.predict(X)
y_km #cada uma das linhas do X
```

```
y_km == 0 # Seleciona as linhas que tem 0 
```

```
Km.cluster_centers_ #Apos o treino ele te diz qual a posicao dos centros.
```

```
X[:,0] # me devolva toda as linhas do inicio ao fim da minha coluna 0
```

```
plt.scatter(
    X[y_km == 0, 0], X[y_km == 0, 1],
    s=50, c='lightgreen',
    marker='s', edgecolor='black',
    label='cluster 1'
)
plt.scatter(
    X[y_km == 1, 0], X[y_km == 1, 1],
    s=50, c='orange',
    marker='o', edgecolor='black',
    label='cluster 2'
)
plt.scatter(
    X[y_km == 2, 0], X[y_km == 2, 1],
    s=50, c='lightblue',
    marker='v', edgecolor='black',
    label='cluster 3'
)
# plot the centroids
plt.scatter(
    Km.cluster_centers_[:, 0], Km.cluster_centers_[:, 1],
    s=250, marker='*',
    c='red', edgecolor='black',
    label='centroids'
)
plt.legend(scatterpoints=1)
plt.grid()
plt.show()
```

```
lista_inertias = []
for i in range(1, 14):
    km = KMeans(  n_clusters=i , init='random', n_init=10, max_iter=300, random_state=0  )
    km.fit( X )
    lista_inertias.append(  km.inertia_  
                          
                          
# plot
plt.plot(range(1, 14), lista_inertias, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```
