```
#Cria um modelo treinado com os dados
modelo_treinado_fake = modelo_regressao.fit(dados_regressao)
```

```
#Verifica o desempenho do modelo
eval_modelo = modelo_treinado_fake.evaluate(dados_regressao)
print(eval_modelo.r2)
print(eval_modelo.rootMeanSquaredError)
```

```
#Carrega o CSV para dentro do sistema
dados_compras = spark.read.csv('Ecommerce_Customers.csv', inferSchema=True, header=True)
```

```
dados_compras.show(5)
```

```
dados_compras.printSchema()
```

```
#Traz a biblioteca que vai preparar os dados para serem aplicados nos algoritmos
from pyspark.ml.feature import VectorAssembler
```

```
#Cria um objeto que pode transformar dados
assembler = VectorAssembler(inputCols=['Avg Session Length', 'Time on App', 'Time on Website'],
                            outputCol='features')
```

```
dados_compras_features = assembler.transform(dados_compras)
```

```
dados_compras_features.show(5)
```

```
dados_compras_features.select('features').show(5)
```

```
dados_compras_features = dados_compras_features.select(['features', 'Yearly Amount Spent'])
```

```
dados_compras_features.show(5)
```

```
#Cria objeto para criar os modelos
lr = LinearRegression(labelCol='Yearly Amount Spent')
```

```
#Dividir os dados
train_data, test_data = dados_compras_features.randomSplit([0.7,0.3], 42)
```

```
#Treina o modelo
modelo_vendas = lr.fit(train_data)
```

```
#Verifica as métricas
modelo_eval = modelo_vendas.evaluate(test_data)
print(modelo_eval.r2)
print(modelo_eval.rootMeanSquaredError)
```

```
#Aplica o modelo para verificar os dados de predição
dados_teste = test_data.select("features")
dados_teste.show(5)
```

```
#Aplica o modelo no conjunto de dados
resultado = modelo_vendas.transform(dados_teste)
```

#Criação de um pipeline de dados para aplicar um modelo de regressão logistica

```
dados_titanic = spark.read.csv('titanic.csv', header=True, inferSchema=True)
```

```
dados_titanic.show(3)
```

```
#Tirar os dados com referência nula
dados_titanic = dados_titanic.na.drop()
```

```
#Divide os dados (treino/teste)
dados_treino, dados_teste = dados_titanic.randomSplit([0.8,0.2], 42)
```

```
#Importa as bibliotecas para aplicar o algoritmo de regressão e construir o pipeline
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.feature import (VectorAssembler, StringIndexer, VectorIndexer, OneHotEncoder)
from pyspark.ml.pipeline import Pipeline
```

```
#Sex -> male, female -> 0,1 e 1,0
indexador_sexo = StringIndexer(inputCol='Sex', outputCol='SexIndex')
```

```
#Criar um vetor da informação do index
classes_sexo = OneHotEncoder(inputCol='SexIndex', outputCol='SexVec')
```

```
indexador_embarque = StringIndexer(inputCol='Embarked', outputCol='EmbarkedIndex')
classes_embarque = OneHotEncoder(inputCol='EmbarkedIndex', outputCol='EmbarkedVec')
```

```
modelo_reg_log = LogisticRegression(featuresCol='features', labelCol='Survived')
```

```
assembler = VectorAssembler(inputCols=['Pclass', 'SexVec', 'EmbarkedVec', 'Age'],
                            outputCol='features')
```

```
#Construir um pipeline para aplicação do modelo
pipeline = Pipeline(stages=[indexador_sexo, indexador_embarque, classes_sexo, classes_embarque,\
                            assembler, modelo_reg_log])
```

```
#Constroi o modelo do pipeline
modelo_reg = pipeline.fit(dados_treino)
```

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator
```

```
resultados = modelo_reg.transform(dados_teste)
```

```
evaluator = BinaryClassificationEvaluator(rawPredictionCol='prediction', labelCol='Survived')
```

```
print(evaluator.evaluate(resultados))
```
