```
#Uso do bibliotecas
import requests

#Pede para o programa fazer um pedido
url = "http://roll.diceapi.com/json/1000d6"

#Realiza a requisicao
resultado = requests.get(url)

print(resultado.text)
```

```
dado_d6 = {}
for dado in resultado.json()["dice"]:
  if dado["value"] in dado_d6.keys():
    dado_d6[dado["value"]] += 1
  else:
    dado_d6[dado["value"]] = 1

print(dado_d6)
```

```
dado_d6 = []
dado_d20 = []
for dado in resultado.json()["dice"]:
  if dado["type"] == 'd20':
    dado_d20.append(dado["value"])
  elif dado["type"] == 'd6':
    dado_d6.append(dado["value"])

print(len(dado_d6))
print(dado_d6)

print(len(dado_d20))
print(dado_d20)
```

```
#Dataframes para analise de dados
import pandas as pd

dados = pd.Series(dado_d6)

print(dados)
```

```
data = {"d6":dado_d6, "d20":dado_d20}
data_frame_dados = pd.DataFrame(data) 
```
